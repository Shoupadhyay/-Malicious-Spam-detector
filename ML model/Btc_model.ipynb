```python
#### we now use random forest ml approach for the same and using the one hot encoder for the categroical data to hash


import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import OneHotEncoder

# load data
df2 = pd.read_csv("/content/drive/MyDrive/concatenated.csv")

# split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df2.drop("label", axis=1), df2["label"], test_size=0.2, random_state=42)

# one-hot encode the categorical features in the training and testing sets
enc = OneHotEncoder(handle_unknown='ignore')
X_train_encoded = enc.fit_transform(X_train)
X_test_encoded = enc.transform(X_test)

# train a random forest classifier on the training set
rfc = RandomForestClassifier(n_estimators=100, random_state=42)
rfc.fit(X_train_encoded, y_train)

# test the accuracy of the classifier on the testing set
y_pred = rfc.predict(X_test_encoded)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# function to classify a BTC address as good or bad
def classify_btc_address(address):
    X = pd.DataFrame({"address": [address]})
    X_encoded = enc.transform(X)
    return  "Illicit" if rfc.predict(X_encoded)[0] == 1 else "Licit"


address = input("Enter a BTC address: ")

# classify the input address
result = classify_btc_address(address)
print(f"The address is {result}.")
```
